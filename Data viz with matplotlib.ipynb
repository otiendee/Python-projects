```python
#converting from a string to timestamp
df.DATE = pd.to_datetime(df.DATE)
```

```python
type(df.DATE)
```

```python
df.head()
```

```python
#pivoting dataframes
reshaped_df = df.pivot(index='DATE', columns='TAG',values='POST')
reshaped_df
```

```python
reshaped_df.shape
```

```python
reshaped_df.head()
```

```python
#dealing with NaN values
reshaped_df.fillna(0, inplace=True)
reshaped_df.head()
```

```python
#no more NaN values
reshaped_df.isna().values.any()
```

```python
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(16,10))
plt.plot(reshaped_df.index,reshaped_df.python,label='python')
plt.plot(reshaped_df.index,reshaped_df.java,label='java')
plt.legend()
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Post', fontsize=14)

```

```python
#visualizing numerous columns
for column in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[column], 
             linewidth=2, label=reshaped_df[column].name)
    plt.legend()
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Number of Post', fontsize=14)
    
   
```

```python
#smoothing out timeseries data
roll_df = reshaped_df.rolling(window=3).mean()
for column in roll_df.columns:
    plt.plot(roll_df.index, reshaped_df[column], 
             linewidth=2, label=roll_df[column].name)
    plt.legend()
   
    
plt.figure(figsize=(16,10))
```
